```python
#保存CSV文件，prod_inst_id，start_time，end_time三个字段缩短成E科学计数法
def num_out(data):
    data = str(data)+'\t'
    return data
 
df['prod_inst_id'] = df['prod_inst_id'].map(num_out)
```
